```python
fig, ax = plt.subplots()
x = [10, 100, 1000, 10000, 100000]

y = [0.002375, 0.044708, 2.86742, 302.985, 32252.4]
ax.plot(x, y, marker='.', label='Bubble sort')

y = [0.000291, 0.004916, 0.358167, 35.4753, 3459.56]
ax.plot(x, y, marker='.', label='Insertion sort')

y = [0.000458, 0.01075, 0.703209, 65.8402, 6285.96]
ax.plot(x, y, marker='.', label='Selection sort')

plt.yscale("log")
plt.xscale("log")
ax.set_title('duration of all sorts')
ax.set_xlabel("number of elements")
ax.set_ylabel("time, ms")
ax.legend()

plt.show()
```
